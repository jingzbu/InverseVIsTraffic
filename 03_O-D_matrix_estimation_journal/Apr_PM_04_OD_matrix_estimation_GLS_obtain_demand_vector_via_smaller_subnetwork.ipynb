```python
od_pair_label_dict_full_ = {}
for key in od_pair_label_dict_full.keys():
    od_pair_label_dict_full_[str(od_pair_label_dict_full[key])] = key
```

```python
od_demand_vector_simplified_dict = {}

for idx in range(len(od_pair_label_dict_full)):
    od_demand_vector_simplified_dict[idx] = 0

for idx in range(len(od_demand_vector_simplified)):
    key = od_pair_label_dict_full_[str(od_pair_label_dict[idx])]
    od_demand_vector_simplified_dict[key] = od_demand_vector_simplified[idx]
```

```python
od_demand_dict_detailed = {}
for i in od_nodes_dict.keys():
    for j in od_nodes_dict.keys():
        if i == j:
            for origi in od_nodes_dict[i]:
                for desti in od_nodes_dict[i]:
                    od_demand_dict_detailed[str(origi) + ',' + str(desti)] = 0
        else:
            for origi in od_nodes_dict[i]:
                for desti in od_nodes_dict[j]:
                    key = od_pair_label_dict_full_[str([i, j])]
                    od_demand_dict_detailed[str(origi) + ',' + str(desti)] = \
                    od_demand_vector_simplified_dict[key]
```
