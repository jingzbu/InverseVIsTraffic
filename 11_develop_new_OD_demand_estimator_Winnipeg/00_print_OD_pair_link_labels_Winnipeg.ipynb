```python
import json

link_length_dict_Winnipeg = {}

length_list = []
with open('../data_traffic_assignment_uni-class/Winnipeg_net.txt', 'r') as f:
    read_data = f.readlines()
    flag = 0
    for row in read_data:
        if ';' in row:
            flag += 1
            if flag > 1:
                length_list.append(float(row.split('\t')[4]))
                
for idx in range(len(length_list)):
    key = str(idx)
    link_length_dict_Winnipeg[key] = length_list[idx]

with open('../temp_files/link_length_dict_Winnipeg.json', 'w') as json_file:
    json.dump(link_length_dict_Winnipeg, json_file)
```

```python
# create O-D pair labels
# create a dictionary mapping O-D pairs to labels

import json

n = 24  # number of nodes

OD_pair_label_dict = {}
OD_pair_label_dict_ = {}

label = 1
for i in range(n + 1)[1:]:
    for j in range(n + 1)[1:]:
        key = (i, j)
        OD_pair_label_dict[str(key)] = label
        OD_pair_label_dict_[str(label)] = key
        label += 1
        
with open('../temp_files/od_pair_label_dict_Winnipeg.json', 'w') as json_file:
    json.dump(OD_pair_label_dict, json_file)
    
with open('../temp_files/od_pair_label_dict__Winnipeg.json', 'w') as json_file:
    json.dump(OD_pair_label_dict_, json_file)
```

```python
# create O-D pair labels
# create a dictionary mapping O-D pairs to labels

import json

n = 24  # number of nodes

OD_pair_label_dict_refined = {}
OD_pair_label_dict_refined_ = {}

label = 1
for i in range(n + 1)[1:]:
    for j in range(n + 1)[1:]:
        if i != j:
            key = (i, j)
            OD_pair_label_dict_refined[str(key)] = label
            OD_pair_label_dict_refined_[str(label)] = key
            label += 1
        
with open('../temp_files/od_pair_label_dict_Winnipeg_refined.json', 'w') as json_file:
    json.dump(OD_pair_label_dict_refined, json_file)
    
with open('../temp_files/od_pair_label_dict__Winnipeg_refined.json', 'w') as json_file:
    json.dump(OD_pair_label_dict_refined_, json_file)
```
