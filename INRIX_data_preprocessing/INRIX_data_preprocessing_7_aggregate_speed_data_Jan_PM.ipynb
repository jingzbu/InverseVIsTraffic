```python
# AM: 7:00 am - 9:00 am
# MD: 11:00 am - 13:00 pm
# PM: 17:00 pm - 19:00 pm
# NT: 21:00 pm - 23:00 pm
```

```python
class TMC_Day_Speed(object):
    def __init__(self, tmc, day, speed, travel_time):
        self.tmc = tmc
        self.day = day
        self.speed = speed
        self.travel_time = travel_time
    def ave_speed(self):
        return sum([self.speed[i] * self.travel_time[i] for i in range(len(self.speed))]) / sum(self.travel_time)
```

```python
data_folder = '/home/jzh/INRIX/All_INRIX_2012_filtered/'
```

```python
tmc_day_speed_dict = {}

import glob


i = 0
for f_name in glob.glob(data_folder + 'Jan_PM/' + '*'):

    i = i + 1
    
    df = load(f_name)

    tmc = list(df.loc[:,'tmc'])[0]
    day = list(df.loc[:,'day'])[0]
    speed = list(df.loc[:,'speed'])
    travel_time = list(df.loc[:,'travel_time'])

    tmc_day_speed = TMC_Day_Speed(tmc, day, speed, travel_time)

    tmc_day_speed_dict[tmc + str(day)] = tmc_day_speed
    
    if i % 500 == 0:
        print "%d TMC-day pairs have been put into the dictionary" %i
```
