```python
%run ../Python_files/util.py

import pandas as pd

tmc_ref_speed_dict =zload('../temp_files/tmc_ref_speed_dict.pkz')

data_folder = '/home/jzh/INRIX/All_INRIX_2012_filtered/'

# load the HDFS for each month
def df(month):
    if month == 'Jan':
        store = pd.HDFStore(data_folder + 'store_df_Jan.h5')
        return store['df_Jan'] 
    elif month == 'Apr':
        store = pd.HDFStore(data_folder + 'store_df_Apr.h5')
        return store['df_Apr'] 
    elif month == 'Jul':
        store = pd.HDFStore(data_folder + 'store_df_Jul.h5')
        return store['df_Jul'] 
    elif month == 'Oct':
        store = pd.HDFStore(data_folder + 'store_df_Oct.h5')
        return store['df_Oct'] 
    else:
        raise IOError("Invalid input; please input another month.")

# obtain the number of days plus 1 for each month
def days(month):
    if month == 'Jan':
        return 32
    elif month == 'Apr':
        return 31
    elif month == 'Jul':
        return 32
    elif month == 'Oct':
        return 32
    else:
        raise IOError("Invalid input; please input another month.")
        
# extract speed data for each tmc, day, hour, minute, corresponding to different periods
def extract_speed(month):
    df_ = df(month)
    i = 0

    for tmc in tmc_ref_speed_dict.keys():
        days_ = days(month)
        for day in range(days_)[1:]:
            i = i + 1
            df_AM_tmc_day = df_.loc[(df_['tmc'] == tmc) & (df_['day'] == day) & \
                                    ((df_['hour'] == 7) | (df_['hour'] == 8))]
            df_MD_tmc_day = df_.loc[(df_['tmc'] == tmc) & (df_['day'] == day) & \
                                    ((df_['hour'] == 11) | (df_['hour'] == 12))]
            df_PM_tmc_day = df_.loc[(df_['tmc'] == tmc) & (df_['day'] == day) & \
                                    ((df_['hour'] == 17) | (df_['hour'] == 18))]
            df_NT_tmc_day = df_.loc[(df_['tmc'] == tmc) & (df_['day'] == day) & \
                                    ((df_['hour'] == 21) | (df_['hour'] == 22))]
            
            df_AM_tmc_day_ = df_AM_tmc_day.sort_values(['hour','minute'])
            df_MD_tmc_day_ = df_MD_tmc_day.sort_values(['hour','minute'])
            df_PM_tmc_day_ = df_PM_tmc_day.sort_values(['hour','minute'])
            df_NT_tmc_day_ = df_NT_tmc_day.sort_values(['hour','minute'])

            pd.DataFrame.to_pickle(df_AM_tmc_day_, data_folder + \
                                   '%s_AM/' %month + 'df_%s_AM_tmc_day_list_%s_%s_%s' %(month, month, tmc, day))
            pd.DataFrame.to_pickle(df_MD_tmc_day_, data_folder + \
                                   '%s_MD/' %month + 'df_%s_MD_tmc_day_list_%s_%s_%s' %(month, month, tmc, day))
            pd.DataFrame.to_pickle(df_PM_tmc_day_, data_folder + \
                                   '%s_PM/' %month + 'df_%s_PM_tmc_day_list_%s_%s_%s' %(month, month, tmc, day))
            pd.DataFrame.to_pickle(df_NT_tmc_day_, data_folder + \
                                   '%s_NT/' %month + 'df_%s_NT_tmc_day_list_%s_%s_%s' %(month, month, tmc, day))
            if i % 701 == 0:
                print("%d df's created" %i)
```

```python
from multiprocessing import Process
    
#create 4 sub-processes to do the work
p1 = Process(target=extract_speed, args=('Jan',))
p1.start()
p1.join()

p2 = Process(target=extract_speed, args=('Apr',))
p2.start()
p2.join()

p3 = Process(target=extract_speed, args=('Jul',))
p3.start()
p3.join()

p4 = Process(target=extract_speed, args=('Oct',))
p4.start()
p4.join()
```
