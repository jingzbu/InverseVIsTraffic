```python
print('probability,' + 'length (miles),' + 'free-flow travel time (hours),' + 'route' + '\n')

for od in range(1, 57):
    for route in od_route_mapping_MA[str(od)]:
        print(str(route_choice_CDC_Apr_PM['(' + str(od) + ',' + str(route) + ')']) + ',' +
              str(route_lens_MA[str(route)]) + ',' + 
              str(route_times_MA[str(route)]) + ',' +
              str(routes_MA[str(route-1)]))
```

```python
od_pairs = []
for i in range(9)[1:]:
    for j in range(9)[1:]:
        if i != j:
            od_pairs.append((i, j))
```

```python
with open('../temp_files/route_choice_analysis_CDC_Apr_PM.csv', 'w') as the_file:    
    the_file.write('Origin,' + 'Destination,' + 'probability,' + 'length (miles),' + 'free-flow travel time (hours),' + 'route' + '\n')
    for od in range(1, 57):
        for route in od_route_mapping_MA[str(od)]:
            the_file.write(str(od_pairs[od-1][0]) + ',' + str(od_pairs[od-1][1]) + ',' 
                           + str(route_choice_CDC_Apr_PM['(' + str(od) + ',' + str(route) + ')']) + ',' +
                  str(route_lens_MA[str(route)]) + ',' + 
                  str(route_times_MA[str(route)]) + ',' +
                  str(routes_MA[str(route-1)]))
```
