```python
# load logit_route_choice_probability_matrix
P = zload('../temp_files/OD_pair_route_incidence_MA.pkz')
P = np.matrix(P)
```

```python
P
```

```python
od_route_incidence_dict = {}

for od_idx in range(np.shape(P)[0]):
    for route_idx in range(np.shape(P)[1]):
        od_route_incidence_dict[(od_idx+1, route_idx+1)] = P[od_idx, route_idx]
```

```python
od_route_incidence_dict
```

```python
od_route_incidence_dict[1,4]
```

```python
odIdx_routeIdx_routeLen_routeChoiceProb_dict = {}

for od_idx in range(np.shape(P)[0]+1)[1:]:
    routeIdx = []
    routeLen = []
    routeChoiceProb = []
    for route_idx in range(np.shape(P)[1]+1)[1:]:
        if od_route_incidence_dict[(od_idx, route_idx)] == 1:
            routeIdx.append(route_idx)
            routeLen.append(route_length_dict_CDC[route_idx])
            routeChoiceProb.append(route_choice_dict_CDC_Apr_PM['(%s,%s)'%(od_idx,route_idx)])
    data = {'routeIdx':routeIdx, 'routeLen':routeLen, 'routeChoiceProb':routeChoiceProb}
    odIdx_routeIdx_routeLen_routeChoiceProb_dict[od_idx] = data
```
